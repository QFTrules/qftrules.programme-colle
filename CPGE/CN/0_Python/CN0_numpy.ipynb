## A) Introduction

NumPy est utilisé pour travailler avec des tableaux. L'objet tableau dans NumPy est appelé ndarray. 

Les tableaux sont très utiles en physique car on peut les manipuler très simplement dans les expressions mathématiques.

Nous pouvons créer un objet NumPy ndarray en utilisant la fonction array() appliquée à une liste.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

print(arr)

print(type(arr))
```

Dans les tableaux, une dimension correspond à un niveau de profondeur du tableau (tableaux imbriqués). 

Nous créons ci-dessous un tableau de dimension 1. La méthode ndim renvoit la dimension d'un tableau.

```python
arr = np.array([1, 2, 3, 4, 5, 6])

print(arr.ndim)
```

On peut obtenir le premier élément du tableau avec la syntaxe suivante.

Nous pouvons prendre la tranche du tableau arr des indices 1 (inclus) à 3 (exclu) de la façon suivante.

```python
arr[1:3]
```

## C) Manipulation mathématique de tableaux

L'intérêt des tableaux NumPy est qu'on peut les manipuler mathématiquement comme des nombres. 

Cela inclut les opérations suivantes : effectuer des additions ou multiplications avec un nombre, 
ou appliquer une fonction au tableau (cos, sin, exp, ln, etc.).

Avec la syntaxe suivante, chaque entrée du tableau est additionné avec le nombre 1.

```python
arr = np.array([1, 2, 3, 4, 5, 6])

print(arr + 1)
print(1 + arr)
```

La même opération effectuée sur une liste renvoit une erreur !

```python
liste = [1, 2, 3, 4, 5, 6]

print(liste + 1)
```

De même, on peut multiplier chaque entrée du tableau avec le nombre 2.

```python
arr = np.array([1, 2, 3, 4, 5, 6])

print(2*arr)
print(arr*2)
```

La même opération effectuée sur une liste renvoie la liste concaténée avec elle-même.

```python
liste = [1, 2, 3, 4, 5, 6]

print(2*liste)
```

## D) Tableaux régulièrement espacés 

Il arrive souvent qu'on ait besoin de créer un tableau constitué d'un nombre N connu de valeurs réelles régulièrement espacées.

La syntaxe pour créer de tels tableaux utilise la fonction linspace. Par exemple, pour créer un tableau comprenant les nombres entiers de 1 à 10, on peut écrire la chose suivante.

```python
N   = 10

arr = np.linspace(1, 10, N)

print(arr)
print(type(arr))
```
