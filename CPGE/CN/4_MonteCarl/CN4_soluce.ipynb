# CN4 - SIMULATION MONTE-CARLO

## A) Introduction

En montage d'optique, on utilise souvent des lentilles comme système de projection. Pour extraire des données quantitatives de l'expérience, il faut donc connaître précisément la distance focale $f'$ de la lentille. On peut pour cela mesurer les positions d'un objet $A$ sur l'axe optique et de son image $A'$ puis appliquer la formule de Descartes

$$f' = \left(\frac{1}{Y} - \frac{1}{X}\right)^{-1}$$

où l'on a noté $X=\overline{OA}$ la distance lentille-objet et $Y=\overline{OA'}$ la distance lentille-image .

Connaissant les incertitudes sur $X$ et $Y$, peut-on déterminer l'incertitude sur $f'$ ?
Dans le TP0 sur les incertitudes, nous avons vus différentes formules de composition des incertitudes. 

- Pour une somme ou une différence $Z=X\pm Y$, on a $\Delta Z = \sqrt{\Delta X^2 + \Delta Y^2}$.

- Pour un produit ou une division $Z=XY$ ou $X/Y$, on a $\frac{\Delta Z}{Z} = \sqrt{(\frac{\Delta X}{X})^2 + (\frac{\Delta Y}{Y})^2}$.

Mais pour la fonction de $X$ et $Y$ ci-dessus, vous ne connaissez pas de formule qui donne l'incertitude sur $f'$ ! On recourt alors à une simulation de type Monte-Carlo pour étudier la variabilité de la grandeur $f'$ sans avoir à composer d'incertitudes.

## B) Préliminaire

>**Question 1**
> 
> Définir en python une fonction `focale` qui retourne $f'$ connaissant $X$ et $Y$.

```python
def focale(X,Y):
    """
    Entrées : distance objet X, distance image Y (float)
    Sortie : distance focale f' (float)
    """
    return 1/(1/Y - 1/X)
```

Un expérimentateur a mesuré les positions d'un objet et d'une image et à trouvé $x=-15$ cm et $y=30$ cm respectivement. En estimant sur quelle plage de positions il voit l'image nette, il a estimé que les incertitudes sur $X$ et $Y$ à $\Delta X = \Delta Y = \Delta = 1$ cm. Pour une simulation Monte-Carlo, nous devons maintenant proposer une distribution de probabilités pour les variables aléatoires $X$ et $Y$. On suppose cette distribution uniforme, centrée sur $x=-15$ cm et $y=30$ cm respectivement, et de demi-largeur égale $\Delta = 1$ cm, selon le code suivant.

```python
""" Initialisations """
x     = -15   # cm, position de l'objet
y     = 30    # cm, position de l'image
Delta = 1     # cm, incertitude-type sur les positions
```

## C) Simulation

La fonction `np.random.uniform` permet de générer un nombre flottant aléatoire tiré d'une distribution uniforme. Les échantillons sont répartis uniformément sur l'intervalle [low, high]. En d'autres termes, toute valeur dans l'intervalle donné a la même probabilité d'être tirée selon la distribution uniforme.

`np.random.uniform(low,high)`

Paramètres :

- `low` : float ou tableau de flottants, optionnel
  Limite inférieure de l'intervalle de sortie. Toutes les valeurs générées seront supérieures ou égales à low. La valeur par défaut est 0.

- `high` : float ou tableau de flottants
  Limite supérieure de l'intervalle de sortie. Toutes les valeurs générées seront inférieures ou égales à high. La limite supérieure peut être incluse dans le tableau de flottants retourné en raison de l'arrondi des nombres à virgule flottante dans l'équation low + (high-low) * random_sample(). La valeur par défaut est 1.0.

```python
""" Importation des bibliothèques """
import numpy as np

""" Initialisations """
N = 100  # nombre de tirages
f = []   # liste de valeurs de f'

""" Tirage de N valeurs de f' """
for i in range(N):
    X = np.random.uniform(x-Delta,x+Delta)  # tirage de X
    Y = np.random.uniform(y-Delta,y+Delta)  # tirage de Y
    f.append(focale(X,Y))                   # calcul de f'
```

La figure comportera un titre et un libellé pour les abscisses. On pourra pour cela utiliser la fonction `matplotlib.pyplot.hist`, dont on donne la documentation ci-dessous.

> `matplotlib.pyplot.hist`

```python
""" Importation des bibliothèques """
import matplotlib.pyplot as plt

""" Affichage de l'histogramme """
plt.close()
plt.hist(f, bins = 40)              # histogramme avec 40 boîtes
plt.title(r"Distribution de f'")    # titre
plt.xlabel(r"f' (cm)")              # libellé des abscisses
plt.show()                          # afficher la figure
```

> **Question 4**
> 
> Déterminer et afficher la valeur moyenne et l'écart-type de la liste f. En déduire le résultat de mesure de $f'$ avec son incertitude-type. On pourra utiliser les fonctions `np.mean` et `np.std`.

```python
fmoy    = np.mean(f)       # valeur moyenne de la distribution de f'
Deltaf  = np.std(f)        # écart-type de la distribution de f'
print("f' = (%.1f ± %.1f) cm"%(fmoy,Deltaf))
```

```python
""" Initialisations """
N = 10000 # nombre de tirages
f = []    # liste de valeurs de f'

""" Tirage de N valeurs de f' """
for i in range(N):
    X = np.random.uniform(x-Delta,x+Delta)  # tirage de X
    Y = np.random.uniform(y-Delta,y+Delta)  # tirage de Y
    f.append(focale(X,Y))                   # calcul de f'

""" Importation des bibliothèques """
import matplotlib.pyplot as plt

""" Affichage de l'histogramme """
plt.close()
plt.hist(f, bins = 40)              # histogramme avec 40 boîtes
plt.title(r"Distribution de f'")    # titre
plt.xlabel(r"f' (cm)")              # libellé des abscisses
plt.show()                          # afficher la figure
```
